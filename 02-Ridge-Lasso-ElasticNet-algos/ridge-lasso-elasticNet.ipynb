# Data Cleaning

```python
## check the missing values
dataset.isnull().sum()
```

```python
 # checking missing values in columns
dataset[dataset.isnull().any(axis=1)]
```

### The dataset is converted into two sets based on Region from 122th index, we can make a new column based on the region 

### 1: "Bejaia Region Dataset"
### 2: "Sidi-Bel Abbes Region Dataset"

### Add new column with region

```python
dataset.loc[:122,"Region"] = 0
dataset.loc[122:,"Region"] = 1
df = dataset
```

```python
df.info()
```

```python
# we see Region column is added and is of type float, we will convert it to int to save memory
df[["Region"]] = df[["Region"]].astype(int)
```

```python
df.info()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# removing the null values
df = df.dropna().reset_index(drop=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
 # remove the 122 row
df =df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
# we find there are some spaces in the column names therefore sometimes we need to remove those
df.columns = df.columns.str.strip()
df.columns
```

```python
# Change the required columns as integer datatype
df[['month','day','year','Temperature','RH','Ws']] = df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

```python
# Changing the other features to float datatype
# By using the generic way
objects = [features for features in df.columns if df[features].dtypes == 'O']
```

```python
objects
```

```python
for i in objects:
    if i != 'Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
# As we got our cleaned dataset , lets save it as another csv file
df.to_csv("Algerian_forest_fires_cleaned.csv", index=False)  
# index=False means we do not want to include the index column in the output file
```
