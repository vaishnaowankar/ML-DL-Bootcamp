```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
california = fetch_california_housing()
```

```python
california.keys() # This line gives what all information this dataset contains
```

```python
california
```

```python
# checking the description of the dataset
print(california.DESCR)
```

```python
# To check what we need to predict
california.target_names
```

```python
# To check the data
print(california.data)
```

```python
# to check target data
print(california.target)
```

```python
# To get the column names
california.feature_names
```

```python
# creating a dataframe
dataset = pd.DataFrame(california.data, columns=california.feature_names)
dataset
```

```python
# the dependent feature is missing and we need to add it
dataset['Price'] = california.target
```

```python
dataset.head()
```

```python
# check the missing values
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
# check the correlation
dataset.corr()
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(dataset.corr(),annot=True)
```

```python
# Dividing data into independent and dependent features
X = dataset.iloc[:,:-1] # independent features
y = dataset.iloc[:,-1]  # dependent features
```

```python
X.head()
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=10)
```

```python
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

```python
# Standardizing the dataset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)   
```

```python
X_train
```

```python
X_test
```

```python
# model training
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor
```

```python
regressor.fit(X_train,y_train)
```

```python
regressor.coef_ # coef will be calculated for all the independent features
```

```python
regressor.intercept_
```

```python
# prediction for test data
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

```python
# finding the performance metrics
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse)
print("MAE:",mae)
print("MSE:",mse)
print("RMSE:",rmse)
```

```python
# R squared and adjusted R squared
score = r2_score(y_test,y_pred)
score
```

```python
# adjusted R squared
1-(1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```

```python
# Assumptions to check the model performance
# 1. Checking the linear relation between y_test and y_pred
plt.scatter(y_test,y_pred)
plt.xlabel("Test original Data")
plt.ylabel("Test Predicted Data")
```

```python
# 2. checking the residuals 
residuals  = y_test - y_pred
```

```python
residuals
```

```python
sns.kdeplot(residuals) # should get normal distribution
```

```python
# 3. The plot of y_pred and residuals should be uniform i.e randomly scattered here and there
plt.scatter(y_pred,residuals) 
```

```python
# In this case we are seeing some relation , hence the accuracy is very low
```

### if we want to take this model and deploy to production we create its pickle file

## Pickling 
Python pickle module is used for serialising and de-serialising a Python object structure. Any object in Python can be pickled so that it can be saved on disk. What pickle does is that it â€œserialisesâ€ the object first before writing it to file. Pickling is a way to convert a python object (list, dict, etc.) into a character stream. The idea is that this character stream contains all the information necessary to reconstruct the object in another python script.


```python
import pickle

pickle.dump(regressor,open('regressor.pkl','wb'))
```

```python
# loading the pickle file
model = pickle.load(open('regressor.pkl','rb'))
```

```python
model.predict(X_test)
```
